### 4.1.1 TOTAL HIT SONGS BY YEAR

### 4.1.3 TOP 10 SONGS BY POPULARITY
<font color="black" size="4.9" face="Times New Roman" style="line-height: 1.3;">This chart shows the top ten most popular songs over 20 years.

<font color="black" size="4.9" face="Times New Roman" style="line-height: 1.3;">Index introduction:
    
<font color="red" size="4.9" face="Times New Roman" style="line-height: 1.3;">Popularity: 
<font color="black" size="4.9" face="Times New Roman" style="line-height: 1.3;">Popularity is an index created by Spotify to measure the popularity of a song. The higher the value, the more popular the song is.

<font color="black" size="4.9" face="Times New Roman" style="line-height: 1.3;">Data analysis: 

<font color="black" size="4.9" face="Times New Roman" style="line-height: 2;">From this chart, Sweater Weather (by The neighbourhood) and Another Love (by Tom Odell) are the most popular songs and have far greater popularity values than the rest of the chart.

</font>

```python
# Select the top 10 songs based on popularity
top_songs = df.nlargest(10, 'popularity')

# Define colors for each artist using different color
colors = plt.cm.tab10.colors
artist_list = top_songs['artist'].unique()
color_dict = dict(zip(artist_list, colors))

# In order to adjust the position of the label that makes it clear to show, writing a for loop to make adjustments for every single label
plt.figure(figsize=(14, 7))
for _, row in top_songs.iterrows():
    plt.scatter(
        x=row['year'], 
        y=row['popularity'], 
        s=row['popularity'],
        color=color_dict[row['artist']],
        alpha=0.9, 
        edgecolors='w',
        linewidth=2
    )
    if row['song'] == 'Sweater Weather':
        plt.text(
            x=row['year'] - 2.5, 
            y=row['popularity'] - 0.1, 
            s=f"{row['song']} by {row['artist']}\n{row['popularity']}",  
            horizontalalignment='center',
            verticalalignment='bottom',
            size=9
        )
    elif row['song'] == 'Wait a Minute!':
            plt.text(
                x=row['year'] - 1.8, 
                y=row['popularity'] - 0.1, 
                s=f"{row['song']} by {row['artist']}\n{row['popularity']}",  
                horizontalalignment='center',
                verticalalignment='bottom',
                size=9
        )

    elif row['song'] == 'The Real Slim Shady':
            plt.text(
                x=row['year']+2, 
                y=row['popularity'] - 0.1, 
                s=f"{row['song']} by {row['artist']}\n{row['popularity']}", 
                horizontalalignment='center',
                verticalalignment='bottom',
                size=9
        )
    elif row['song'] == 'lovely (with Khalid)':
            plt.text(
                x=row['year']-1.1, 
                y=row['popularity']+0.1, 
                s=f"{row['song']} by {row['artist']}\n{row['popularity']}",  
                horizontalalignment='center',
                verticalalignment='bottom',
                size=9
        )        
    else:   
        plt.text(
            x=row['year'], 
            y=row['popularity'] +0.1, 
            s=f"{row['song']} by {row['artist']}\n{row['popularity']}",  
            horizontalalignment='center',
            verticalalignment='bottom',
            size=9
        )

# Set the title and labels for the plot

plt.title('Top 10 Songs by Popularity', fontsize=19, color='green')
plt.xlabel('Year',fontsize=12)
plt.ylabel('Popularity',fontsize=12)

# Add gridlines to the plot
plt.grid(True, linestyle='--', alpha=0.5, which='both', axis='both')

# Set the x-axis and y_axis ticks 

plt.xticks(df['year'].unique()) 
plt.yticks(range(0, 101, 10))

#Add legend entries for each artist
for artist in artist_list:
    plt.scatter([], [], c=color_dict[artist], alpha=0.6, s=100, label=artist)
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='Artist')


plt.show()
```

### 4.1.4 TOP 10 SINGERS WITH MOST HITS
<font color="black" size="4.9" face="Times New Roman" style="line-height: 1.3;">This chart shows the top ten most <font color="black" size="4.9" face="Times New Roman" style="line-height: 1.3;">popular songs over 20 years.
<font color="black" size="4.9" face="Times New Roman" style="line-height: 2;">This chart statistics the artists who have had the most number of hits in this time period

<font color="black" size="4.9" face="Times New Roman" style="line-height: 1.3;">Data analysis: 


<font color="black" size="4.9" face="Times New Roman" style="line-height: 2;">Behind this chart, it can be noticed that Rihanna was the artist with the most number of hits, and only Drake, Calvin Harris, and Eminem had more than 20 hits among the rest.
</font>

```python
# Group the dataframe by artist and count the number of songs for each artist
artist_hits = df.groupby('artist')['song'].count().sort_values(ascending=False)

# Select the top 10 artists with the most hits
top_10_artists = artist_hits.nlargest(10)

# Generate colors for the treemap based on the hit counts
mini, maxi = top_10_artists.min(), top_10_artists.max()
colors = [plt.cm.Greens((value - mini) / (maxi - mini)) for value in top_10_artists]

# Create the figure and label
plt.figure(figsize=(12, 8))
labels = [f"{artist}\n{hits}" for artist, hits in zip(top_10_artists.index, top_10_artists)]
squarify.plot(
    sizes=top_10_artists,
    color=colors,
    alpha=0.7,
    label=labels,
    text_kwargs={'fontsize': 10, 'fontweight': 'bold'}
)

plt.title('Top 10 Singers with Most Hits', fontsize=19,color='green')
plt.axis('off')
plt.show()
```

### 4.1.6 LONGEST SONG  VS  SHORTEST SONG 
<font color="black" size="4.9" face="Times New Roman" style="line-height: 1.3;">The following table showing the longest and shortest in the given period.
</font>






```python
#Transforming duration from million  format into min:sec format 
def ms_to_minsec(ms):
    sec = ms / 1000
    return f"{int(sec // 60)}:{int(sec % 60)}"
transformation = df.loc[:, ['artist','song','duration_ms','year']]
transformation['min:sec'] = transformation['duration_ms'].apply(ms_to_minsec)
```

```python
# Longest song
transformation[transformation.duration_ms == transformation.duration_ms.max()]
```

```python
# Shortest song
transformation[transformation.duration_ms == transformation.duration_ms.min()]
```

### 4.2.1 Hit song style analysis(Mode)
<font color="black" size="4.9" face="Times New Roman" style="line-height: 1.3;">Index introduction:

<font color="red" size="4.9" face="Times New Roman" style="line-height: 1.3;">Major Mode: 
<font color="black" size="4.9" face="Times New Roman" style="line-height: 1.3;">The Major Mode is based on the major scale. It is commonly used to express <font color="red" size="4.9" face="Times New Roman" style="line-height: 1.3;">happiness, grandeur, optimism, and positivity </font>in music.

    
<font color="red" size="4.9" face="Times New Roman" style="line-height: 1.3;">Minor Mode:
<font color="black" size="4.9" face="Times New Roman" style="line-height: 1.3;">The Minor Mode is based on the minor scale. It is often used to convey <font color="red" size="4.9" face="Times New Roman" style="line-height: 1.3;">sadness, longing,  introspection, and melancholy </font>in music.
    
    
<font color="black" size="4.9" face="Times New Roman" style="line-height: 1.3;">Data analysis: 
    
<font color="black" size="4.9" face="Times New Roman" style="line-height: 2;">It can be noticed that, with the exception of 2017, the line representing the major mode is always above the minor mode, indicating that in the past 20 years of hits, More songs expressed positive attitudes rather than negative attitudes in the given period.
</font>

```python
#Grouping dataframe and creating the data
mode_counts = df.groupby(['year', 'mode']).size().unstack(fill_value=0)
plt.figure(figsize=(14, 6))


# Plot each mode's count over time

for mode in mode_counts.columns:
    plt.plot(mode_counts.index, mode_counts[mode], marker='o', label=f'Mode {mode}')
    for x, y in zip(mode_counts.index, mode_counts[mode]):
        if mode == 'mode 1':
            plt.text(x, y, str(y), ha='center', va='top')
        else:
            plt.text(x, y, str(y), ha='center', va='bottom')

            
            
            
# Plot elements setting            
years = range(2000, 2020)
plt.xticks(mode_counts.index, [f'{year}' for year in years])

plt.title('Major and Minor Mode Over Time',color='green', fontsize=19)
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='Mode')

plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

### 4.2.2 Hit song style analysis(Danceability)
<font color="black" size="4.9" face="Times New Roman" style="line-height: 1.3;">Index introduction:

<font color="red" size="4.9" face="Times New Roman" style="line-height: 1.3;">Danceability: 
<font color="black" size="4.9" face="Times New Roman" style="line-height: 2;">Danceability describes how suitable a track is for dance. Songs with high danceability values usually mean a more positive musical style, faster tempo.

    

    
<font color="black" size="4.9" face="Times New Roman" style="line-height: 1.3;">Line graph description:
    
<font color="black" size="4.9" face="Times New Roman" style="line-height: 2;">From 2000 to 2015, there was no obvious high or low differentiation, with little variance between these years. However, an increasing trend of this index can be seen in the last 5 years, reaching its peak in 2018.
</font>

```python
#Grouping the dataframe and Creating bar chart 
danceability_mean = df.groupby('year')['danceability'].mean()
plt.figure(figsize=(14, 7))
plt.plot(danceability_mean.index, danceability_mean, marker='o', linestyle='-', color='#FF9999')

# Add value labels to each data point
for year, mean_value in danceability_mean.items():
    plt.text(year, mean_value+0.01, f"{mean_value:.2f}", ha='center', va='bottom')

    
# Plot elements setting            
plt.title('Danceability Mean Value', fontsize=19, color='green')
plt.xlabel('Year')
plt.ylabel('Mean Danceability')
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(danceability_mean.index)
plt.yticks([i/100 for i in range(35, 80, 5)])

plt.show()
```

### 4.2.3 Hit song style analysis(Valence)
<font color="black" size="4.9" face="Times New Roman" style="line-height: 1.3;">Index introduction:

<font color="red" size="4.9" face="Times New Roman" style="line-height: 1.3;">Valence: 
<font color="black" size="4.9" face="Times New Roman" style="line-height: 2;">A measure from 0.0 to 1.0 describes the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).

    

    
<font color="black" size="4.9" face="Times New Roman" style="line-height: 1;">Line graph description:
    
<font color="black" size="4.9" face="Times New Roman" style="line-height: 2;">For the first eighteen years, this chart demonstrates an initially fluctuating but continuously declining trend. From 2018, a positive upward trend can be seen in 2019. But overall, it is a generally declining trend in the whole given period.
</font>

```python
#Grouping the dataframe and Creating bar chart 
valence_mean = df.groupby('year')['valence'].mean()
plt.figure(figsize=(14, 7))
plt.plot(valence_mean.index, valence_mean, marker='o', linestyle='-', color='#CC9933')



# Add value labels to each data point
for year, mean_value in valence_mean.items():
    plt.text(year, mean_value+0.015, f"{mean_value:.2f}", ha='center', va='bottom')


# Plot elements setting            
plt.title('Valence Mean Value', fontsize=19, color='green')
plt.xlabel('Year')
plt.ylabel('Mean Valence')
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(valence_mean.index)
plt.yticks([i/100 for i in range(35, 80, 5)])

plt.show()
```

### 4.2.4 Hit song style analysis(Energy)
<font color="black" size="4.9" face="Times New Roman" style="line-height: 1.3;">Index introduction:

<font color="red" size="4.9" face="Times New Roman" style="line-height: 1.3;">Energy: 
<font color="black" size="4.9" face="Times New Roman" style="line-height: 2;">Energy is an index used to describe the rhythmic strength of music. It is a scale from 0.0 to 1.0, representing a perceived measure of intensity and activity. The higher the value of musical energy, the stronger the rhythmic changes. Drum beats are clearer and more enthusiastic, and the mood is positively outgoing.
    

    
<font color="black" size="4.9" face="Times New Roman" style="line-height: 1.3;">Line graph description:
    
<font color="black" size="4.9" face="Times New Roman" style="line-height: 2;">Since 2000, the average value of energy has fluctuated upwards, peaking at 0.77 in 2012, and then fluctuating downwards until 2019，Overall, this graph shows that the average energy value of the music has remained high over the 20-year period, despite the ups and downs, pointing to a trend in the overall dynamics and intensity of the music's changes.
</font>

```python
#Grouping the dataframe and Creating bar chart 
energy_mean = df.groupby('year')['energy'].mean()
plt.figure(figsize=(14, 7))
plt.plot(energy_mean.index, energy_mean, marker='o', linestyle='-', color='#00CCCC')


# Add value labels to each data point
for year, mean_value in energy_mean.items():
    plt.text(year, mean_value-0.025, f"{mean_value:.2f}", ha='center', va='bottom')

    
# Plot elements setting            
plt.title('Energy Mean Value', fontsize=19, color='green')
plt.xlabel('Year')
plt.ylabel('Mean Valence')
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(energy_mean.index)
plt.yticks([i/100 for i in range(35, 80, 5)])

plt.show()
```

<font color="black" size="4.9" face="Times New Roman" style="line-height: 2;">
    
    
I think that even though these indexes <font color="red" size="4.9" face="Times New Roman" style="line-height: 2;">(Energy, Danceability, Valence)</font>are trying to represent different emphases, in general, they all express positive or negative emotions about song styles, so I will next put the line graphs of these three indexes on a single graph for comparative data analysis.

```python
# Calculate and plot the mean values for each year and features
means = df.groupby('year')[['valence', 'danceability', 'energy']].mean()

plt.figure(figsize=(14, 8))

plt.plot(means.index, means['valence'], marker='o', label='Valence')
plt.plot(means.index, means['danceability'], marker='o', label='Danceability')
plt.plot(means.index, means['energy'], marker='o', label='Energy')

# Writing for loop to adjust the label position that may cause confusing understanding
for year in means.index:
    plt.text(year, means.at[year, 'valence']-0.015, f"{means.at[year, 'valence']:.2f}", ha='center', va='bottom')
    
    if year == 2016:
        adjusted_y_position = means.at[year, 'energy'] +0.01
        plt.text(year, adjusted_y_position, f"{means.at[year, 'energy']:.2f}", ha='center', va='bottom')
        adjusted_y_position = means.at[year, 'danceability']- 0.02 
        plt.text(year, adjusted_y_position, f"{means.at[year, 'danceability']:.2f}", ha='center', va='bottom')
    elif year == 2017:
        adjusted_y_position = means.at[year, 'energy']-0.02
        plt.text(year, adjusted_y_position, f"{means.at[year, 'energy']:.2f}", ha='center', va='bottom')
        adjusted_y_position = means.at[year, 'danceability']+ 0.01 
        plt.text(year, adjusted_y_position, f"{means.at[year, 'danceability']:.2f}", ha='center', va='bottom')

    else:
        plt.text(year, means.at[year, 'energy'], f"{means.at[year, 'energy']:.2f}", ha='center', va='bottom')
        plt.text(year, means.at[year, 'danceability'], f"{means.at[year, 'danceability']:.2f}", ha='center', va='bottom')



# Plot elements setting            
plt.title('Comparison between Valence, Danceability, and Energy', fontsize=19, color='green')
plt.xlabel('Year')
plt.ylabel('Mean Value')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

plt.xticks(means.index)

plt.tight_layout()
plt.show()


```

<font color="black" size="4.9" face="Times New Roman" style="line-height: 2;">
    
    
I use the value of 0.5 as a dividing line, <font color="red" size="4.9" face="Times New Roman" style="line-height: 2;">greater than 0.5 means that the energy of the song is higher, it sounds more dynamic and strong</font>, and the emotional attitude of the song is more positive, <font color="red" size="4.9" face="Times New Roman" style="line-height: 2;">while less than 0.5 means that the energy of the song is lower, it sounds more soothing and calming,</font> and the emotional attitude of the song is inclined to be more negative, e.g. longing, sadness, introspection, etc. 
<font color="black" size="4.9" face="Times New Roman" style="line-height: 2;">

From this comparison chart, we can see that the average values of  <font color="red" size="4.9" face="Times New Roman" style="line-height: 2;">Energy </font>and <font color="red" size="4.9" face="Times New Roman" style="line-height: 2;">Danceability </font>are both greater than 0.5 in each of the 20 years, so if we use these two indicators to determine what kind of emotions Spotify's top songs in the past 20 years are inclined to, <font color="red" size="4.9" face="Times New Roman" style="line-height: 2;">we can actually find that most of the songs will be positively inclined.</font> However, if we use the <font color="red" size="4.9" face="Times New Roman" style="line-height: 2;">Valence </font>index, we can see that there have been a number of years where the average has been below 0.5. <font color="red" size="4.9" face="Times New Roman" style="line-height: 2;">The overall sentiment of the songs would be more negative.
